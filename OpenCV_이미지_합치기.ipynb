```
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

image_1 = cv2.imread('flower.png')
image_2 = np.array(Image.open('background.jpg'))

fg_h,fg_w,_ = image_1.shape

bg_h, bg_w ,_ = image_2.shape

print(image_1.shape)
print(image_2.shape)

image_2 = cv2.resize(image_2,dsize=(fg_w,int(fg_w*bg_h / bg_w))) #가로길이 맞추기

bg_h, bg_w ,_ = image_2.shape

margin = (bg_h - fg_h) // 2   #위아래 늘리니깐 2번

print(margin)
if margin >0:
  image_2 = image_2[margin:-margin,:,:]
else:
  image_2 = cv2.copyMakeBorder(image_2,top =abs(margin),bottom=abs(margin),left=0,
                    right=0,borderType=cv2.BORDER_REPLICATE)
image_2 = cv2.resize(image_2,(1000,600))
image_1 = cv2.imwrite('content/flower_1.jpg',image_1)
image_1.Image.open('flower_1.jpg')
image_1.save('flower_1.png')

```

```

result = cv2.add(image_2, image_1)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```
